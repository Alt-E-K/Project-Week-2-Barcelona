# ACCIDENTS AND VICTIMS BY STREET

```python
#Import libraries
import pandas as pd
import numpy as np
```

```python
#Import dataframe of streets of Barcelona and storing raw data as copy
streets_raw = pd.read_csv('../datasets/0401230100_BCN_Nomenclàtor.csv')
streets = streets_raw.copy()
streets.head()
```

```python
#Created a new column with a concatenation of 3 columns: TIPUS_VIA, PARTICULES, NOM
streets['Complete_name'] = streets['TIPUS_VIA'].str.capitalize().str.cat([streets['PARTICULES'], streets['NOM']], sep= ' ', na_rep = '')

streets_clean = streets[['CODI', 'Complete_name']]
streets_clean.columns = ['Codi_carrer', 'Complete_name']

streets_clean
```

```python
#Imported the dataframe acc_clean from the other Notebook.
acc_clean = pd.read_csv('../datasets/acc_clean.csv', index_col = 0)
acc_clean.head()
acc_clean
```

```python
# Here we get a merge (left join) between acc_clean and the streets, in order to get exactly the street's name for each accident.
streets_accidents = acc_clean.merge(streets_clean, how = 'left', on = 'Codi_carrer' )
streets_accidents.head()
```

```python
# Now we get the grouby by street name, in a similar way as the notebook about accidents by day/hour or month.
streets_accidents_grouped=streets_accidents.groupby('Complete_name').aggregate({'Numero_expedient': 'count', 'Numero_victimes': 'sum', 
                                                      'Numero_vehicles_implicats': 'sum', 'Numero_lesionats_lleus':'sum', 
                                                      'Numero_lesionats_greus':'sum', 'Numero_morts':'sum'})
streets_accidents_grouped.columns = ['Accidents', 'Total_Victims', 'Total_Vehicles', 'Minor_Injuries', 'Severe_Injuries', 'Deaths']
streets_accidents_grouped.index.name= 'Street_name'

streets_accidents_grouped_perc = streets_accidents_grouped.apply(lambda x: round (100 * x / x.sum(),2))
streets_accidents_grouped_perc.columns = ['%Accidents', '%Total_Victims', '%Total_Vehicles', '%Minor_Injuries', '%Severe_Injuries', '%Deaths']


streets_accidents_grouped = pd.concat([streets_accidents_grouped,streets_accidents_grouped_perc], axis =1)
streets_accidents_grouped = streets_accidents_grouped[['Accidents', '%Accidents', 'Total_Victims', '%Total_Victims', 'Total_Vehicles', '%Total_Vehicles', 'Minor_Injuries',
       '%Minor_Injuries', 'Severe_Injuries', '%Severe_Injuries', 'Deaths', '%Deaths']]
streets_accidents_grouped
```

### Accidents by street

```python
# Get the values of the resulting table sorted by number of accidents.
streets_accidents_grouped[['Accidents','%Accidents']].sort_values('Accidents', ascending=False).head(30)
```

### Total victims by street

```python
streets_accidents_grouped[['Total_Victims','%Total_Victims']].sort_values('Total_Victims', ascending=False).head(20)
```

### Severe injuries by street

```python
streets_accidents_grouped[['Severe_Injuries','%Severe_Injuries']].sort_values('Severe_Injuries', ascending=False).head(20)
```

### Deaths by street

```python
streets_accidents_grouped[['Deaths','%Deaths']].sort_values('Deaths', ascending=False).head(20)
```
